# SALARY PREDICTION

# Based on years of experience

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset from a CSV file
data = pd.read_csv('Salary_Data_2.csv')

# Define the predictors and response variable
X = data[['YearsExperience']]
y = data['Salary']

# Assign feature names to X
X.columns = ['YearsExperience']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Print the model coefficients
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)

# Plot the data and the linear regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Predict the salary for a new observation
new_observation = [[10]]
predicted_salary = model.predict(new_observation)
print('Predicted salary:', predicted_salary)
```

# Based on the gender

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the dataset from a CSV file
data = pd.read_csv('Salary_Data_2.csv')

# Convert the gender variable to binary variables using one-hot encoding
data = pd.get_dummies(data, columns=['Gender'])

# Define the predictors and response variable
X = data[['Gender_0', 'Gender_1']]
y = data['Salary']

# Create a linear regression model
GenderModel = LinearRegression()

# Fit the model to the data
GenderModel.fit(X, y)

# Print the model coefficients
print('Intercept:', GenderModel.intercept_)
print('Coefficients:', GenderModel.coef_)

# Plot the data and the linear regression line
plt.scatter(X, y)
plt.plot(X, GenderModel.predict(X), color='blue')
plt.title('Salary vs Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()

# Predict the salary for a new observation
new_observation = [[0, 1]]  # Gender = Male
predicted_salary = model.predict(new_observation)
print('Predicted salary:', predicted_salary)
```

# Based on Department

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the dataset from a CSV file
data = pd.read_csv('Salary_Data_3.csv')

# Convert the gender variable to binary variables using one-hot encoding
data = pd.get_dummies(data, columns=['Department'])

# Define the predictors and response variable
X = data[['Department_0', 'Department_1', 'Department_2', 'Department_3']]
y = data['Salary']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Print the model coefficients
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)

# Predict the salary for a new observation
new_observation = [[0, 1, 0, 0]]  # Gender = Male
predicted_salary = model.predict(new_observation)
print('Predicted salary:', predicted_salary)
```
